```python
# 1  Loading and Preprocessing
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data  
y = iris.target 
```

```python

iris_df = pd.DataFrame(X, columns=iris.feature_names)
```

```python
# 2 Clustering Algorithm Implementation 
# Apply KMeans clustering with 3 clusters 
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
# the cluster labels and centroids
kmeans_labels = kmeans.labels_
kmeans_centroids = kmeans.cluster_centers_
```

```python
# visualization of clusters 
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis')
plt.scatter(kmeans_centroids[:, 0], kmeans_centroids[:, 1], s=200, c='red', marker='X', label="Centroids")
plt.title("KMeans Clustering (Iris Dataset)")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend()
plt.show()
```

```python
# Apply Agglomerative Hierarchical Clustering (3 clusters)
hierarchical = AgglomerativeClustering(n_clusters=3)
hierarchical_labels = hierarchical.fit_predict(X)
```

```python
# Visualize the clusters (use first two features for 2D visualization)
plt.scatter(X[:, 0], X[:, 1], c=hierarchical_labels, cmap='viridis')
plt.title("Hierarchical Clustering (Iris Dataset)")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```
